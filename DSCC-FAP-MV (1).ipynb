## Importing Libraries

```python
import yfinance as yf
import pandas as pd
import statistics
import matplotlib.pyplot as plt
```

## Extracting Data using Yahoo Finance's yfinance Module

```python
class FinancialDataDownloader:
    def __init__(self, start_date, end_date):
        self.start_date = start_date
        self.end_date = end_date

    def download_stock_data(self, symbol):
        data = yf.download(symbol, start=self.start_date, end=self.end_date)
        return data

class DataDisplay:
    def __init__(self, data):
        self.data = data

    def display_tabular_format(self):
        df = pd.DataFrame(self.data)
        print(df)

# Main program
if __name__ == '__main__':
    downloader = FinancialDataDownloader('2023-01-01', '2023-03-31')

    # Download data for Tesla
    tesla_data = downloader.download_stock_data('TSLA')
    print("Tesla Stock Data:")
    display_tesla = DataDisplay(tesla_data)
    display_tesla.display_tabular_format()

    # Download data for Ford
    ford_data = downloader.download_stock_data('F')
    print("Ford Stock Data:")
    display_ford = DataDisplay(ford_data)
    display_ford.display_tabular_format()
```

## Saving DataFrame to CSV File

```python
class FinancialDataDownloader:
    def __init__(self, start_date, end_date):
        self.start_date = start_date
        self.end_date = end_date

    def download_stock_data(self, symbol):
        data = yf.download(symbol, start=self.start_date, end=self.end_date)
        return data

    def save_data_to_csv(self, data, filename):
        data.to_csv(filename)

class DataDisplay:
    def __init__(self, data):
        self.data = data

    def display_tabular_format(self):
        df = pd.DataFrame(self.data)
        print(df)

# Main program
if __name__ == '__main__':
    downloader = FinancialDataDownloader('2023-01-01', '2023-03-31')

    # Download data for Tesla
    tesla_data = downloader.download_stock_data('TSLA')
    downloader.save_data_to_csv(tesla_data, 'tesla_data.csv')

    # Download data for Ford
    ford_data = downloader.download_stock_data('F')
    downloader.save_data_to_csv(ford_data, 'ford_data.csv')

    print("Data saved to CSV files.")
```

```python
#Store closing prices from the stock data into separate variables
tesla_close_prices = df_tesla["Close"]
ford_close_prices = df_ford["Close"]
```

```python
# Calculate statistics for tesla closing prices
tesla_stats = {
    "Min": min(tesla_close_prices),
    "Max": max(tesla_close_prices),
    "Range": max(tesla_close_prices) - min(tesla_close_prices),
    "Median": statistics.median(tesla_close_prices),
    "Mean": statistics.mean(tesla_close_prices),
    "Variance": statistics.variance(tesla_close_prices),
    "Standard Deviation": statistics.stdev(tesla_close_prices)
}
```

```python
# Calculate statistics for ford closing prices
ford_stats = {
    "Min": min(ford_close_prices),
    "Max": max(ford_close_prices),
    "Range": max(ford_close_prices) - min(ford_close_prices),
    "Median": statistics.median(ford_close_prices),
    "Mean": statistics.mean(ford_close_prices),
    "Variance": statistics.variance(ford_close_prices),
    "Standard Deviation": statistics.stdev(ford_close_prices)
}
```

```python
# Create a pandas DataFrame to display the statistics
df = pd.DataFrame([tesla_stats, ford_stats], index=["Tesla", "Ford"])

# Display the DataFrame
df.head()
```

```python
# Plotting the time series
plt.figure(figsize=(12, 6))
plt.plot(df_tesla['Date'], tesla_close_prices, label="Tesla")
plt.plot(df_ford['Date'], ford_close_prices, label="Ford")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.title("Stock Prices Over Time")
plt.legend()
plt.show()
```
